```python
import pandas as pd
import math
```

```python
df = pd.read_csv('data.csv')
```

```python
df.head(3)
```

```python
df.columns
```

```python
#colunas q da pra binarizar em >60 ou n
to_create = ['Overall', 'Potential','LS', 'ST', 'RS', 'LW', 'LF', 'CF', 'RF', 'RW', 'LAM', 'CAM',
       'RAM', 'LM', 'LCM', 'CM', 'RCM', 'RM', 'LWB', 'LDM', 'CDM', 'RDM',
       'RWB', 'LB', 'LCB', 'CB', 'RCB', 'RB', 'Crossing', 'Finishing',
       'HeadingAccuracy','ShortPassing', 'Volleys', 'Dribbling', 'Curve',
       'FKAccuracy', 'LongPassing', 'BallControl', 'Acceleration',
       'SprintSpeed', 'Agility', 'Reactions', 'Balance', 'ShotPower',
       'Jumping', 'Stamina', 'Strength', 'LongShots', 'Aggression',
       'Interceptions', 'Positioning', 'Vision', 'Penalties', 'Composure',
       'Marking', 'StandingTackle', 'SlidingTackle', 'GKDiving', 'GKHandling',
       'GKKicking', 'GKPositioning', 'GKReflexes' ]
```

```python
#criando as colunas
for column in to_create:
    array = []
    for i in range(len(df)):
        if df[column].iloc[i] > 60:
            array.append(1)
        else:
            array.append(0)
    new_column_name = column + '>60'
    df[new_column_name] = array
```

```python
df.head(3)
```

```python
yet_to_use = df.columns.tolist()
#removendo as que tem >60
yet_to_use = yet_to_use[:83]
#removendo as que ja foram usadas pra criar as >60
for i in to_create:
    yet_to_use.remove(i)
#tirando a primeira que era 'Unnamed: 0'
yet_to_use = yet_to_use[1:]
print(yet_to_use)
```

```python
#dummy ones
dummy = yet_to_use
for i in ['ID', 'Name', 'Value', 'Wage', 'Height', 'Weight', 'Special', 'Joined', 'Contract Valid Until', 'Release Clause']:
    try:
        dummy.remove(i)
    except:
        continue
dummy
```

```python
#creating dummys
for column in dummy:
    uniques = df[column].unique().tolist()
    #isso aqui nao funciona!
    for item in uniques:
        if type(item) == float:
            if math.isnan(item):
                uniques.remove(item)
        
    for item in uniques:
        array = []
        for i in range(len(df)):
            if df[column].iloc[i] == item:
                array.append(1)
            else:
                array.append(0)
        
        new_column_name = column + '_' + str(item)
        df[new_column_name] = array                 
```

```python
df.head(3)
```

```python
texto_to_csv = df.to_csv(index = False)
f = open('data_com_dummy.csv', 'w')
f.write(texto_to_csv)
f.close()
```

```python
nan_vec = []
for i in dummy:
    valor = False
    for item in df[i].unique().tolist():
        if type(item) == float:
            if math.isnan(item):
                valor = True
                break
    nan_vec.append(valor)
        
```

```python
print(dict(zip(dummy, nan_vec)))
```
